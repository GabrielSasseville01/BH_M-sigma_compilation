```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy

# Values for mu and c
mu_values = [2, 7]
c_values = [1.0, 4]

# Generate x values
x = np.linspace(-10, 10, 1000)

# Plot the PDF for different mu and c values
plt.figure(figsize=(10, 6))
for mu, c in zip(mu_values, c_values):
    plt.plot(x, levy.pdf(-x, -mu, c), label=f'mu={mu}, c={c}')

plt.title('Lévy Distribution PDF with Different Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(False)
plt.show()
```
